```python
6. If a dictionary is stored in spam, what is the difference 
  between the expressions 'cat' in spam and 'cat' in spam.values()?

Answer: -
    
The 'cat' in spam expression will check for 'cat' in keys of the 
dictionary spam  

Whereas 'cat' in spam.values() expression will check for 'cat' in values
of the dictionary spam
```
